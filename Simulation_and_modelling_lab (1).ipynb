**Project Title:**"Analyzing Exam Performance: Online vs Offline Students"

<br>


**Dataset:**
Letâ€™s create a small fictional dataset with exam scores for two groups:

Group A: 10 students who studied online

Group B: 10 students who studied offline

<br>

**Step-by-Step Approach:**
Create the dataset using Python.

1. Visualize the score distributions (optional but adds value).

2. Perform a Two Sample T-Test (Independent T-Test).

3. Interpret the result (p-value).

```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data: Exam scores out of 100
online_scores = [72, 75, 78, 84, 70, 69, 71, 73, 76, 74]
offline_scores = [80, 85, 79, 82, 81, 83, 78, 86, 84, 80]

# Visualize the data (Optional but good)
sns.histplot(online_scores, color='blue', label='Online', kde=True)
sns.histplot(offline_scores, color='green', label='Offline', kde=True)
plt.title('Distribution of Exam Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Perform Two Sample T-Test (Independent T-Test)
t_stat, p_value = ttest_ind(online_scores, offline_scores)

# Print Results
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in scores.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```
